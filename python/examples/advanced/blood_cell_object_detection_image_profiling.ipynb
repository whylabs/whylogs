```python
import getpass
import os
from datetime import datetime
from typing import Dict

from whylogs.api.writer.whylabs import WhyLabsWriter

# set your org-id here - should be something like "org-xxxx"
os.environ["WHYLABS_DEFAULT_ORG_ID"] = "org-XXXX"

# set your datased_id (or model_id) here - should be something like "model-xxxx"
os.environ["WHYLABS_DEFAULT_DATASET_ID"] = "model-XXXX"

# set your API key here
print("Enter your WhyLabs API key")
os.environ["WHYLABS_API_KEY"] = getpass.getpass()
print("Using API Key ID: ", os.environ["WHYLABS_API_KEY"][0:10])
writer = WhyLabsWriter()
reference_writer = WhyLabsWriter()
```

```python
from datasets import load_dataset

ds = load_dataset("keremberke/blood-cell-object-detection", name="full")
example = ds['train'][0]
```

```python
from PIL import Image, ImageDraw
import whylogs as why
from whylogs.extras.image_metric import log_image
import pandas as pd
import numpy as np

row_count = len(ds['train']['image'])
views = []
batch_size = 10
max_row = min(batch_size, row_count) # just for debugging, limit to 10 to view
for row_id in range(max_row):
    image = ds['train']['image'][row_id]
    im_data = np.asarray(image)

    # tile image into quadrants
    M = im_data.shape[0]//2
    N = im_data.shape[1]//2
    tiles = [im_data[x:x+M,y:y+N] for x in range(0,im_data.shape[0],M) for y in range(0,im_data.shape[1],N)]
    single_image = log_image(image)
    tiled_results = log_image([Image.fromarray(tile) for tile in tiles])

    im_draw = image.copy()
    image_draw = ImageDraw.Draw(im_draw)

    objects = ds['train']['objects'][row_id]
    bbox_count = len(objects['bbox'])
    results = single_image.merge(tiled_results)
    results = results.merge(why.log({"image bbox_count": bbox_count}))
    area_sum = 0
    for index in range(bbox_count):
        bbox = objects['bbox'][index]
        xmin, ymin, width, height = bbox
        category = objects['category'][index]
        area = objects['area'][index]
        area_sum = area_sum + area
        results = results.merge(why.log({"image category": category}))
        results = results.merge(why.log({"image area": area}))
        outline = None
        if category == 0:
            outline = "#DDA0DD"
        elif category == 1:
            outline = "#FF0000"
        elif category == 2:
            outline = "#ffffff"
        image_draw.rectangle(xy=[xmin, ymin, xmin+width, ymin+height], outline=outline)
    # for demo debug purposes, here's the image and the bbox with color coded annotation
    display(im_draw)
    results = results.merge(why.log({"image area_sum": area_sum}))

    image_id = ds['train']['image_id'][row_id]
    reference_name = f"ref_image_id_{image_id}"
    results = results.merge(why.log({"outliers": reference_name}))
    view = results.view()
    print(f"image_id={image_id}")
    print(view.get_columns().keys())
    print(view.get_column('image').to_summary_dict())
    print(f"image bbox_count {view.get_column('image bbox_count').to_summary_dict()}")
    print(f"image area {view.get_column('image area').to_summary_dict()}")
    print(f"image area_sum {view.get_column('image area_sum').to_summary_dict()}")
    print(f"image category {view.get_column('image category').to_summary_dict()}")
    writer.write(view)
    reference_writer.option(reference_profile_name=reference_name).write(view)
    views.append(view)

```

```python
from datetime import timedelta, datetime, timezone
import random

now = datetime.now(timezone.utc)
view.dataset_timestamp = now

# simulate a week of data
for i in range(1, 8):
    view = random.choice(views)
    for _ in range(3):
        view = view.merge(random.choice(views))
    view.dataset_timestamp = now - timedelta(days=i)
    print(f"Writing test profile for {view.dataset_timestamp} to WhyLabs")
    writer.write(view)
```
